#  알고리즘 기초
## 알고리즘 이란 ?
- 어떠한 문제를 해결하기 위해 정해 놓은 일련의 절차
- 어떠한 경우에도 실행 결과가 똑같이 나오는 것

```python
# 세 정수를 입력받아 최댓값 구하기

print('세 정수의 최댓값을 구합니다.')
a = int(input('정수 a의 값을 입력하세요: '))
b = int(input('정수 b의 값을 입력하세요: '))
c = int(input('정수 c의 값을 입력하세요: '))

maximum = a
if b > maximum: maximum = b
if c > maximum: maximum = c

print(f'최댓값은 {maximum}입니다.')
```

```python
print(int('19')) # 10진수 문자열을 10진수 정수형으로 변환 (기본)
print(int('0b110', 2)) # 2진수 문자열을 10진수 정수형으로 변환
print(int('0o75', 8)) # 8진수 문자열을 10진수 정수형으로 변환
print(int('13',10)) # 10진수 문자열을 10진수 정수형으로 변환
print(int('0x3F',16)) # 16진수 문자열을 10진수 정수형으로 변환
print(float('3.14')) # 문자열을 실수형으로 변환
```

```python
def max3(a,b,c) :
  maximum = a
  if b > maximum: maximum = b
  if c > maximum: maximum = c
  return maximum

print(f'maximum3(3,2,1) = {max3(3,2,1)}')
print(f'maximum3(3,2,2) = {max3(3,2,2)}')
print(f'maximum3(3,2,3) = {max3(3,2,3)}')
print(f'maximum3(3,1,2) = {max3(3,1,2)}')
print(f'maximum3(2,1,3) = {max3(2,1,3)}')
print(f'maximum3(3,3,2) = {max3(3,3,2)}')
print(f'maximum3(3,3,3) = {max3(3,3,3)}')
print(f'maximum3(2,2,3) = {max3(2,2,3)}')
print(f'maximum3(2,3,1) = {max3(2,3,1)}')
print(f'maximum3(2,3,2) = {max3(2,3,2)}')
print(f'maximum3(1,3,2) = {max3(1,3,2)}')
print(f'maximum3(2,3,3) = {max3(2,3,3)}')
print(f'maximum3(1,2,3) = {max3(1,2,3)}')
```

```python
# 중앙값 구하기
def med3(a,b,c) :
  if a >= b :
    if b >= c :
      return b
    elif a <= c :
      return a
  elif a > c :
    return a
  elif b > c :
    return c
  else :
    return b


a = int(input('정수 a의 값을 입력하세요: '))
b = int(input('정수 b의 값을 입력하세요: '))
c = int(input('정수 c의 값을 입력하세요: '))

print(f'중앙값은 {med3(a,b,c)} 입니다.')
```

```python
def med3(a,b,c):
    if (a >= b and a <= c) or (a <= b and a >= c):
        return a
    elif (b >= a and b <= c) or (b <= a and b >= c):
        return b
    else:
        return c

a = int(input('정수 a의 값을 입력하세요: '))
b = int(input('정수 b의 값을 입력하세요: '))
c = int(input('정수 c의 값을 입력하세요: '))

print(f'중앙값은 {med3(a,b,c)} 입니다.')
```
